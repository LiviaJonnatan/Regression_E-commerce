## Categorical - Integer Encoding

```python
df.Avatar.unique()
```

```python
# integer encoding
numerical = 0
for avatar in df.Avatar.unique():
    numerical = numerical + 1
    df.loc[df['Avatar'] == avatar, 'avatar_int_encoding'] = numerical
```

```python
df[['Avatar', 'avatar_int_encoding']]
```

```python
# binary encoding
numerical = 0
for avatar in df.Avatar.unique():
    numerical = numerical + 1
    df[f'avatar_{numerical}'] = 0
    df.loc[df['Avatar'] == avatar, f'avatar_{numerical}'] = 1
```

```python
df.head()
```

```python
# one hot encoding
for avatar in df.Avatar.unique():
    df[f'avatar_{avatar}'] = 0
    df.loc[df['Avatar'] == avatar, f'avatar_{avatar}'] = 1
```
